```python
filt=(df1['Model']=='d3q27q27_cm_cht_OutFlowNeumann_IBB')
```

```python
speed1=df1.loc[filt]['X']*df1.loc[filt]['Y']*df1.loc[filt]['Z']/df1.loc[filt]['Time']
```

```python
speed1;
```

```python
plt.plot(df1.loc[filt]['Devices'], speed1, 'x')
plt.title("Weak scaling")
plt.xlabel("Number of GPU")
plt.ylabel("Number of elemnts per second")
plt.xticks(np.arange(0, 12, 1))
```

```python
filt2=(df1['Model']=='d3q27q7_cm_cht_OutFlowNeumann_IBB')
```

```python
speed2=df1.loc[filt2]['X']*df1.loc[filt2]['Y']*df1.loc[filt2]['Z']/df1.loc[filt2]['Time']
```

```python
plt.plot(df1.loc[filt2]['Devices'], speed2, 'x')
plt.title("Weak scaling")
plt.xlabel("Number of GPU")
plt.ylabel("Number of elemnts per second")
plt.xticks(np.arange(0, 12, 1))
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, constrained_layout=True)
#fig.figsize=(4,12)
#plt.figure(figsize=(200,500))
fig.suptitle('Weak Scaling', fontsize=25)
ax1.plot(df1.loc[filt]['Devices'], speed1, 'x')
ax2.plot(df1.loc[filt2]['Devices'], speed2, 'x')
#fig.tight_layout(pad=2.0)
#plt.setp(ax2, xlabel='Number of GPU')

ax1.set_xlabel('Number of GPU', fontsize=18)
ax1.set_ylabel('Number of elements per second', fontsize=18)
ax2.set_xlabel('Number of GPU', fontsize=18)
ax2.set_ylabel('Number of elements per second', fontsize=18)
#ax1.set_title('Title', fontsize=14)
ax1.set_xticks(np.arange(0, 13, 1))
ax2.set_xticks(np.arange(0, 13, 1))
ax1.grid(True)
ax2.grid(True)
fig.savefig('fig1.png', dpi = 600)
```
